# 🌍 Earth Engine: AIAD Weather Forecasting Platform

Welcome to the **Advanced Integrated Aerospace Data (AIAD)** Weather forecasting project. 

This platform has evolved through 5 distinct versions, growing from basic data preprocessing to an advanced, multimodal AI reasoning system integrating Google Gemini.

---

## 🚀 Platform Evolution (Versions 1 - 5)

Explore the journey of the AIAD Weather platform:

### [Version 1: Foundation](./Version1) (Archived)
- Initial proof of concept integrating basic satellite imagery.

### [Version 2: Data Preprocessing](./Version2/readme.md)
- Advanced cloud isolation and HSV thresholding.
- Generative Adversarial Network (GAN) modeling for future frame synthesis.

### [Version 3: Visual Analytics](./Version3/index.html)
- Simple HTML/CSS frontend for model validation.
- Time-series GAN optimization and hourly/daily prediction outputs.

### [Version 4: Enterprise Architecture](./Version4/README.md)

### [Version 5: Intelligence Integrated](./Version5/README.md) 🌟 **(CURRENT)**
